```python
df_result = pd.read_pickle('results_Cora.pkl')

ax = sns.lineplot(data=df_result, x='eps', y='acc', hue='conf')
handles, labels = ax.get_legend_handles_labels()
ax.legend(handles=handles[1:], labels=labels[1:])
ax.set_xlim(0,10)
plt.savefig('figures/cora.pdf', bbox_inches='tight')
```
